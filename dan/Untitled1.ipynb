```python
import matplotlib.pyplot as plt
from numpy.random import exponential

from numpy.random import exponential
data = exponential(5,20)
plt.plot(data, color='red', label='Название графика')  # color='red' - опционально, можно color='#ff0000'
plt.title('Результат', fontdict={'fontsize':20})  # Название графика, размер шрифта заголовка
plt.xlabel('Ось Х')
plt.ylabel('Ось Y')
plt.legend()  # Выводит лейблы графиков
plt.xticks(range(0, 20, 4))  # Выводить по оси х отметки с 0 до 20 с интервалом 4
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))  # Рисует 6 графиков - 2 строкиб 3 колонки, figsize - размер графика в дюймах, axes - список списков холстов
for row, row_axes in enumerate(axes):
    for column, ax in enumerate(row_axes):
        ax.plot(exponential(column, 20))  # ax - конкретный холст, по оси y - количество колонок
        ax.set_title('Холст колонка {} строка {}'.format(row+1, column+1))
fig.tight_layout()  # Делает изображение таким, что бы туда поместились всё
fig.savefig('2x3.png')
```
